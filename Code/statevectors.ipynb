```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
```

```python
bell = QuantumCircuit(2)
bell.h(0)
bell.cx(0, 1)
bell.draw(output='mpl')
```

```python
bell_state = Statevector(data=bell)
bell_state.draw(output='latex')
```
